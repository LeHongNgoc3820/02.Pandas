## 1. Đọc/ghi dữ liệu
Định dạng tập tin: *.csv, *.xlsx, *.hdf, ...

Nơi lưu trữ:
+ Máy tính cục bộ: \data\<Tập tin data>
+ Internet: địa chỉ URL

### 1.1 Đọc file csv
Thao tác:  
`pd.read_csv(đường_dẫn_tên_tập_tin, sep = dấu_phân_cách, index_col = index_cột, nrows = số_lượng_dòng, header = None - Nếu không có dòng tiêu đề, usecols = [index_cột, ...], encoding = 'utf-8)`  

Kết quả trả về: DataFrame hoặc TextParser

**Ghi dữ liệu:**  

Thao tác: 

`df.to_csv(đường_dẫn_tên_tập_tin, sep = ',', na_rep = '', columns=None, header=True, mode = 'w', encoding = 'utf-8', decimal='.')`

Với decimal: khai báo phần thập phân là '.' hay ','.

### 1.2 Đọc file excel

Cài thư viện: `pip install xlrd`  

**Đọc file chỉ có 1 sheet:**  

Thao tác:  

`pd.read_excel(đường_dẫn_tên_tập_tin, encoding='utf-8', index_col = index_cột, nrows = số lượng dòng...)`  

Kết quả trả về: DataFrame hoặc TextParser.

**Đọc file có nhiều sheet:**

Thao tác:

`xlsx = pd.ExcelFile()` và `pd.read_excel(xlsx, "Tên sheet")`

Kết quả trả về: DataFrame hoặc TextParser.

### 1.3 Ghi file

`df.to_excel(đường_dẫn_tên_tập_tin, tên_sheet)`

## 2. Làm sạch dữ liệu
**Dữ liệu ban đầu rất lộn xộn bởi vì:**
+ Thiếu giá trị
+ Có các giá trị ngoại lại
+ Có các giá trị không hợp lệ (ví dụ giá trị âm cho tuổi hay cho cân nặng)
+ Chứa các giá trị NaN (np.NaN), không có giá trị (None)

**Xử lí các vấn đề về dữ liệu:**
+ Thay thế giá trị
+ Điền bằng dữ liệu phía trên hoặc phái dưới cho những giá trị NaN
+ Loại bỏ các field
+ Nội suy giá trị

**Các phương thức:**
+ Thay thế dữ liệu, dùng:   

`df.replace(giá_trị_cũ, giá_trị_mới)`

+ Điền bằng dữ liệu phía trên hoặc phía dưới, dùng: 

`df.fillna(method='ffill')` hoặc `df.fillna(method='backfill')`

+ Xoá bỏ dòng dữ liệu có chứa NaN, dùng:  

`df.dropna(axis=0 hoặc 1)`

+ Thực hiện nối suy tuyến tính, dùng:  

`df.interpolate()`

## 5. Gộp dữ liệu
**Thao tác:** 

`pd.concat([df1, df2, ...])`

**Kết dữ liệu:**

`pd.concat([df1, df2, ...], axis = 1, join = 'inner')`

**Trong đó:**
+ axis = 1: lấy theo cột (0 là lấy theo dòng).
+ join = 'inner': loại bỏ hết dòng data null (outer: lấy cả dòng null)

**Nối dữ liệu, dùng:**

`df1.append(df2)`

**Kết dữ liệu:**

`pd.merge(df1, df2, how='inner')`

## 6. Phương thức thao tác trên chuỗi
**Tách dữ liệu chuỗi thành list, dùng:**

`df['Tên cột].str.split('Ký tự tách')`

**Tìm chuỗi có nằm trong chuỗi hay không:**

`df["Tên cột"].str.contains('chuỗi')`

**Thay chuỗi bằng chuỗi:**

`df['Tên cột'].str.replace('Chuỗi cũ', 'chuỗi mới')`

**Tìm chuỗi đầu tiên thoã regular:**

`df["Tên cột"].str.extract('RE')`
## 7. Phương thức thao tác trên Timestamps
Kiểu dữ liệu Timestamps:
+ Kiểu dữ liệu chung: datatime64[ns]
+ Chuyển dữ liệu int64 timestamps sang dữ liệu DataTime của Python

Để chuyển qua dữ liệu DataTime, dùng:

`pd.to_datatime(tên_cột_timestamp, units = 's')`

Sắp xếp dữ liệu theo trật tự thời gian, dùng:

`df.sort_values(by='tên_cột', ascending = true)`
